```python
def rmsle(p, a):
    sum = 0
    for i in  range(0,len(p)):
        sum += math.pow(math.log(float(p[i]) + 1) - math.log(float(a[i]) + 1),2)
    sum = sum * (1/len(p))
    return float(math.sqrt(sum))
```
